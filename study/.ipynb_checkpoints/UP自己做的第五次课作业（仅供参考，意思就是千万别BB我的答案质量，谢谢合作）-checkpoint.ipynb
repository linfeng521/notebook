（6）数据离散化，按照价格将酒店分为3个等级,0-500为C，500-1000为B,大于1000为A，列名设置为“价格等级”。

```python
df['价格等级'] = pd.cut(df['价格'], [0,500,1000,np.inf], labels=['C', 'B', 'A'])
df
```

（7）获取评分均值最高和最低的地区的数据，分别使用append和concat方法将获取的两个数据集合并。

```python
s_sorted = df['评分'].groupby(df['地区']).mean().sort_values()
highest = df[df['地区']==s_sorted.index[-1]]
lowest = df[df['地区']==s_sorted.index[0]]
```

```python
highest.append(lowest)
```

```python
pd.concat([highest, lowest])
```

（8）数据离散化，按照评分人数将酒店平均分为3个等级，三个等级的酒店数量尽量保持一致。评分人数最多的为A，最少的为C。列名设置为“热门等级”。

```python
bins = np.percentile(df['评分人数'],[0,33,67,100])
df['热门等级'] = pd.cut(df['评分人数'], bins, labels = ['C','B','A'])
df.loc[df['评分人数']==1,'热门等级'] = 'C' # 1 的都是 nan, 只能暂时用这种方法修补
df
```

（9）选出评分人数为A，价格也为A的酒店数据，计算其平均评分。

```python
df[(df.热门等级 == 'A') & (df.价格等级 == 'A')]['评分'].mean()
```

（10）取价格最高的5个酒店的数据，使用stack和unstack方法实现dataframe和Series之间的转换。

```python
top5 = df.sort_values(by='价格', ascending=False)[:5]
top5
```

```python
top5_s = top5.stack()
top5_s
```

```python
top_df = top5_s.unstack()
top_df
```
